```
#Playfair cipher
import numpy as np
plaintext="instruments"
#plaintext="me"
key='monarchy'

def keygrid(key):
  l=[]
  alp=[]
  for i in range(26):
    if i!=9 and (chr(97+i) not in key):
      alp.append(chr(97+i))
  #return alp
  for i in key:
    l.append(i)
  for i in alp:
    l.append(i)
  
  l=np.array(l).reshape(5,5)
  return l

def splitp(plain):
  slist=[] #to store split list of digraphs

  if len(plain)%2 != 0:
    plain+='z'
  
  for i in range(0,len(plain),2):
    slist.append(plain[i]+plain[i+1])
  return slist

def encrypt(plain,k,s):
  cipher=[]
  c1=''
  c2=''
  for i in s:
    
    a,b=np.where(k==i[0]) #row and column index of first element
    c,d=np.where(k==i[1]) #row and column index of second element
    #print(a,b,c,d)
    if b[0] == d[0]: #if both in the same column
      c1= k[(a[0]+1)%5][b[0]]
      c2= k[(c[0]+1)%5][d[0]]
    elif a[0] == c[0]: #if both in the same column
      c1= k[a[0]][(b[0]+1)%5]
      c2= k[c[0]][(d[0]+1)%5]
    else:            #if none of above
      c1= k[a[0]][d[0]]
      c2= k[c[0]][b[0]]
  
    cipher.append(c1+c2)
  return "".join(cipher)

def decrypt(cipher,k,s):
  #print(s)
  decry=[]
  d1=''
  d2=''
  for i in s:
    
    a,b=np.where(k==i[0]) #row and column index of first element
    c,d=np.where(k==i[1]) #row and column index of second element
    #print(a,b,c,d)
    if b[0] == d[0]: #if both in the same column
      d1= k[(a[0]-1)%5][b[0]]
      d2= k[(c[0]-1)%5][d[0]]
    elif a[0] == c[0]: #if both in the same column
      d1= k[a[0]][(b[0]-1)%5]
      d2= k[c[0]][(d[0]-1)%5]
    else:            #if none of above
      d1= k[a[0]][d[0]]
      d2= k[c[0]][b[0]]
  
    decry.append(d1+d2)
  return "".join(decry)


k=keygrid(key)
#print(k)
s=splitp(plaintext)
print("Key text: ",key)
print("Plain text: ",plaintext)
cipher=encrypt(plaintext,k,s)
print("Cipher text: ",cipher)
decryt=decrypt(cipher,k,splitp(cipher))
if len(plaintext)%2!=0:
  decryt=decryt[:-1]
print("Decrypted text: ",decryt)
```
